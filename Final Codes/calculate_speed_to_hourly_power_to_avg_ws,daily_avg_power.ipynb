```python
import pandas as pd
```

```python
df1 = pd.read_csv( 'Chandpur_hourly_speed.csv')
```

```python
df1.info()
```

```python
df1['WS50M'].head(20)
```

```python
import numpy as np
```

```python
def calculate_power(ws50):
  if ws50 < 3 :
    return 0
  elif ws50 <= 15:
    x = ws50
    y = 0.0017*x**6 - 0.0762*x**5 + 1.1519*x**4 - 7.4713*x**3 + 29.871*x**2 - 59.961*x + 45.075
    return np.round(y,2)
  elif ws50 > 25:
      return 0
  else:
      return 1000

df1['Power'] = df1['WS50M'].apply(calculate_power)
```
