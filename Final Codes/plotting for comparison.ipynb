```python
import pandas as pd

# Read the CSV file
data = pd.read_csv('/content/Sylhet_daily_avg_power.csv')

# Extract the necessary columns
days = data['Day2017'].tolist()
wind_speeds = data['WS2017'].tolist()

# Plotting the graph
import matplotlib.pyplot as plt

plt.plot(days, wind_speeds, linestyle='-', color='b')
plt.xlabel('Day')
plt.ylabel('Wind Speed')
plt.title('Parkay 2017')
plt.grid(False)
fig = plt.gcf()  # Get the current figure
fig.set_size_inches(18,6)  # Set the width and height of the figure
plt.yticks(range(0, 22, 2))   # Set the lower limit of the y-axis to 0
plt.xlim(0)
# Displaying the graph
plt.show()
```
